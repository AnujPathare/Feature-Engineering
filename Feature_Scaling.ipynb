# Min-Max scaling to transform the values in a dataset to a range of -1 to 1

X-scaled = ((X  min(X)) / (max(X)  min(X))) * (ba) + a

```python
import numpy as np
import pandas as pd

dataset = pd.DataFrame({
    'Income': [20000, 35000, 101010, 47000, 540000, 66666]
})

def custom_minmaxscaler(data, a, b):
  x_min = np.min(data)
  x_max = np.max(data)
  scaled_data = ((data - x_min) / (x_max - x_min)) * (b-a) + (a)
  return scaled_data

scaled_data = custom_minmaxscaler(dataset, a=-1, b=1)
scaled_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, Normalizer
import numpy as np

# Create a sample dataset
np.random.seed(42)
data = {'Feature1': np.random.randint(0, 100, 40),
        'Feature2': np.random.randint(0, 100, 40)}

df = pd.DataFrame(data)

# Initialize MinMaxScaler for Min-Max Scaling
minmax_scaler = MinMaxScaler()

# Initialize Normalizer for Unit Vector Scaling
unit_vector_scaler = Normalizer()

# Apply Min-Max Scaling
df[['Feature1_MinMaxScaled', 'Feature2_MinMaxScaled']] = minmax_scaler.fit_transform(df[['Feature1', 'Feature2']])

# Apply Unit Vector Scaling
df[['Feature1_UnitVectorScaled', 'Feature2_UnitVectorScaled']] = unit_vector_scaler.fit_transform(df[['Feature1', 'Feature2']])

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original Data
axes[0].scatter(df['Feature1'], df['Feature2'], label='Original Data')
axes[0].set_title('Original Data')
axes[0].set_xlabel('Feature1')
axes[0].set_ylabel('Feature2')
axes[0].legend()

# Min-Max Scaled Data
axes[1].scatter(df['Feature1_MinMaxScaled'], df['Feature2_MinMaxScaled'], label='Min-Max Scaled Data', color='orange')
axes[1].set_title('Min-Max Scaled Data')
axes[1].set_xlabel('Feature1')
axes[1].set_ylabel('Feature2')
axes[1].legend()

# Unit Vector Scaled Data
axes[2].scatter(df['Feature1_UnitVectorScaled'], df['Feature2_UnitVectorScaled'], label='Unit Vector Scaled Data', color='green')
axes[2].set_title('Unit Vector Scaled Data')
axes[2].set_xlabel('Feature1')
axes[2].set_ylabel('Feature2')
axes[2].legend()

plt.tight_layout()
plt.show()
```

# PCA

## A general approach to decide on the number of principal components to retain

- Standardize Features:

  - Standardize the features (subtract the mean and divide by the standard deviation) to ensure that they are on a similar scale.
  - Standardizing features is necessary, since, if the variables have different scales, those with larger scales will dominate the principal components, leading to biased results.

- Apply PCA:

  - Apply PCA to the standardized dataset.

- Analyze Explained Variance:

  - Examine the explained variance ratio for each principal component.

- Plot Cumulative Explained Variance:

  - Plot the cumulative explained variance to visualize how much variance is retained as the number of components increases.

- Select Number of Components:

  - Choose the number of components that captures a sufficient amount of variance. A common threshold is often set (e.g., 95% or 99% of total variance).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Create a sample 2D dataset
np.random.seed(42)
data = np.random.randn(100, 2)  # 100 samples, 2 features

# Apply PCA
pca = PCA(n_components=1)
data_pca = pca.fit_transform(data)

# Plot original and PCA-transformed data
plt.figure(figsize=(8, 4))

# Original Data
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], alpha=0.8)
plt.title('Original Data')

# PCA-transformed Data
plt.subplot(1, 2, 2)
plt.scatter(data_pca, np.zeros_like(data_pca), alpha=0.8)
plt.title('PCA-transformed Data')

plt.tight_layout()
plt.show()
```
